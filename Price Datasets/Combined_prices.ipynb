```python
# import necessary libraries
import pandas as pd
import os
import numpy as np
```

```python
path = os.getcwd()
# print(path)
#nake a list to append file
prices_df = []

for filename in os.listdir(path):
    if filename.endswith('.csv'):
        df = pd.read_csv(filename)
        Name = filename.split(".")[0]   #extraxt the name of the place
        df["Place"] = Name    # add name column for further analyses
        prices_df.append(df)  #append the list

        
```

```python
prices_df
```

```python
# get combined df
combined_prices_df = pd.concat(prices_df,axis = 0)
    
```

```python
combined_prices_df
combined_prices_df.dtypes # to ensure we dont run into the errors while doing visualization
```

```python
#save the df
combined_prices_df.to_csv("combined_prices.csv",index=False)
```
