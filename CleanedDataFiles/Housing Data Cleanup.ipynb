```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import os
```

```python
rental_data_to_load = "urban-rental-markets-2021.xlsx"
```

```python
# Read Rental Data File and store into Pandas DataFrames
rental_data = pd.read_excel(rental_data_to_load, skiprows=4, skipfooter=5)
```

```python
rental_data.head()
```

```python
#Delete empty columns using theie index
del_rental_data = rental_data.drop(rental_data.columns[[5,7,9,11,13]], axis = 1)
```

```python
del_rental_data.head()
```

```python
del_rental_data.dtypes
```

```python
#Rename column labels
rename_rental_data = del_rental_data.rename(
    {"Census Subdivision \n Subdivision de recensement" : "Census Subdivision",
     "Dwelling Type \n Type de logement": "Dwelling Type",
     "Bachelor \n Studios" : "Bachelor",
     "1 Bedroom \n 1 chambre" : "1 Bedroom",
    "2 Bedroom \n 2 chambres" : "2 Bedroom",
    "3 Bedroom + \n 3 chambres +" : "3 Bedroom +",
    "Total \n Tous les log." : "Total"}, axis=1)
```

```python
rename_rental_data
```

```python
rename_rental_data.dtypes
```

```python
#remove rows with NaN

del_rental_unit_data = rename_rental_data.dropna()
```

```python
del_rental_unit_data
```

```python
#Replace -- & ** with zero
del_rental_unit_data = del_rental_unit_data.replace({"--":"0","**":"0"})
del_rental_unit_data.head()
```

```python
# Remove '$' from rental unit  & total columns before  data type conversion
del_rental_unit_data["Bachelor"] = del_rental_unit_data["Bachelor"].str.replace("$", "",regex=True)
del_rental_unit_data["1 Bedroom"] = del_rental_unit_data["1 Bedroom"].str.replace("$", "",regex=True)
del_rental_unit_data["2 Bedroom"] = del_rental_unit_data["2 Bedroom"].str.replace("$", "",regex=True)
del_rental_unit_data["3 Bedroom +"] = del_rental_unit_data["3 Bedroom +"].str.replace("$", "",regex=True)
del_rental_unit_data["Total"] = del_rental_unit_data["Total"].str.replace("$", "",regex=True)
```

```python
del_rental_unit_data.head(20)
```

```python
#Convert rental unit  & total column data type from str to int
del_rental_unit_data[["Bachelor","1 Bedroom","2 Bedroom","3 Bedroom +","Total"]] = del_rental_unit_data[["Bachelor","1 Bedroom","2 Bedroom","3 Bedroom +","Total"]].apply(pd.to_numeric)
```

```python
del_rental_unit_data.dtypes
```

```python
#remove French from values in Province & Dwellimg Type columns

del_rental_unit_data["Province"] = del_rental_unit_data["Province"].str.split('/').str[0]
del_rental_unit_data["Dwelling Type"] = del_rental_unit_data["Dwelling Type"].str.split('/').str[0]
```

```python
del_rental_unit_data

```

```python
#remove rows with all zero values, i,e, Total column is zero
del2_rental_unit_data = del_rental_unit_data[del_rental_unit_data["Total"] > 0]
```

```python
del2_rental_unit_data
```

```python
#Remove rows with reporting totals for Centre and Subdivision or Dwelling Type
del3_rental_unit_data = del2_rental_unit_data[(del2_rental_unit_data["Dwelling Type"] != "Total")] 
del3_rental_unit_data = del3_rental_unit_data[(del3_rental_unit_data["Census Subdivision"] != "Total")] 
```

```python
del3_rental_unit_data
```

```python
#Write DataFrame back to an excel file
del3_rental_unit_data.to_excel("urban-rental-markets-2021-Clean.xlsx", index=False)
```
