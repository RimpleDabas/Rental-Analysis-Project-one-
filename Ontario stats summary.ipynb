```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
```

```python
#Reading the files
path_rental = "Resources/Average Rents_Concat.csv"
# path_Number_of_Units = pd.re("Resources/Number_of_Units_Concat.csv")
rental_df = pd.read_csv(path_rental)
rental_df
```

```python
province_stats = rental_df.groupby(['Province'])[["Bachelor ",'1 Bedroom', '2 Bedroom', '3 Bedroom +',"Total "]].agg(['mean', 'var', 'std', 'sem'])
province_stats.round(2)
```

```python
stats = province_stats.to_excel("Statistcs summary.xlsx",
             sheet_name='Ontario') 
```

```python
#Create a funtion to extract based on the province and year 
def data_set(filepath,Province,Year,column):
    get_df = pd.read_csv(filepath)
    df = get_df.loc[(rental_df["Year"] == Year )& (get_df["Province"] == Province)]
    df_Province = df[(df[column]!=0)]
    return df_Province
```

```python
#Check if the function works fine and get the Ont.province
Year = 2022
column = "1 Bedroom"
df_Ontario = data_set(path_rental, "Ont.",Year,column)
df_Ontario
```

```python
#define Provinces name in the list
Province = ['Ont.']
Province 
```

```python
# Using the aggregation method, to produce the summary statistics in a single line for all bedrooms in Ontario
Centre_df= df_Ontario.groupby(['Centre'])[["Bachelor ",'1 Bedroom', '2 Bedroom', '3 Bedroom +',"Total "]].agg(['mean', 'var', 'std', 'sem'])
Centre_df.round(2)
```

```python
df_Toronto = df_Ontario[(df_Ontario["Centre"]=="Toronto")]
Zone_df= df_Toronto.groupby(['Zone'])[["Bachelor ",'1 Bedroom', '2 Bedroom', '3 Bedroom +',"Total "]].agg(['mean', 'var', 'std', 'sem'])
Zone_df.round(2)
```

```python
with pd.ExcelWriter("Statistcs summary.xlsx",
                    mode='a') as writer:  
    Zone_df.to_excel(writer, sheet_name='Toronto')
```
