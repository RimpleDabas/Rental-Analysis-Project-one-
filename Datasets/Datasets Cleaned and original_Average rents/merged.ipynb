```python
import pandas as pd
import numpy as np
```

```python
filepath_2018 = "Average_rents_Clean_2018.csv"
filepath_2019 = "Average_rents_Clean_2019.csv"
filepath_2020 = "Average_rents_Clean_2020.csv"
filepath_2021 = "Average_rents_Clean_2021.csv"
filepath_2022 = "Average_rents_Clean_2022.csv"
```

```python
df_2018 = pd.read_csv(filepath_2018)
df_2019 = pd.read_csv(filepath_2019)
df_2020 = pd.read_csv(filepath_2020)
df_2021 = pd.read_csv(filepath_2021)
df_2022 = pd.read_csv(filepath_2022)
```

```python
df_2018_wy = df_2018.loc[:, df_2018.columns != "Year"]
df_2019_wy = df_2019.loc[:, df_2019.columns != "Year"]
df_2020_wy = df_2020.loc[:, df_2020.columns != "Year"]
df_2021_wy = df_2021.loc[:, df_2021.columns != "Year"]
df_2022_wy = df_2022.loc[:, df_2022.columns != "Year"]
```

```python
df_18_19 = pd.merge(df_2018_wy,df_2019_wy, how = "outer", on =[ "Province","Centre","Zone","Neighbourhood","Dwelling Type "], suffixes = (" 18" ," 19"))
```

```python
df_18_19
```

```python
df_20 = pd.merge(df_18_19,df_2020_wy, how = "outer", on =[ "Province","Centre","Zone","Neighbourhood","Dwelling Type "])
```

```python
df_20
```

```python
df_21 = pd.merge(df_20,df_2021_wy, how = "outer", on =[ "Province","Centre","Zone","Neighbourhood","Dwelling Type "], suffixes = (" 20"," 21"))
```

```python
df_21
```

```python
df_22 = pd.merge(df_21,df_2022_wy, how = "outer", on =[ "Province","Centre","Zone","Neighbourhood","Dwelling Type "])
```

```python
df_22
```

```python
df_22.columns
```

```python
df_combined = df_22.rename(columns = {'Bachelor ':'Bachelor  22' , 
                                      '1 Bedroom': '1 Bedroom 22' , 
                                      '2 Bedroom': '2 Bedroom 22',
                                      '3 Bedroom +' : '3 Bedroom + 22' , 'Total ': 'Total  22'})
```

```python
df_combined = df_combined.fillna(0)
```

```python
df_combined.to_csv("Rents_merged_clean.csv",index=False)
```
