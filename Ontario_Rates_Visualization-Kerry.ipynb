#Ontario Rental Rates
    Analyse rental rates within Onatrio for:
    1. All municipalities (Centres) in 2022
    2. All regions within the GTA in 2022
    3. 5 year trend (2018-2022) for rental rates within the GTA,using the top 5 most expensive 
       regions for 2022.


```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Rental data filepaths
rental_rates_path = "Resources/Average Rents_Concat.csv"
```

```python
# Read the rental, vacancy and units data
rental_data = pd.read_csv(rental_rates_path)
```

```python
rental_data.head()
```

#    Bar Graphs for rental rates

```python
#Get total (combined) average rental rate by year
rent_2018 = rental_data[(rental_data["Year"] == 2018)]
rent_2019 = rental_data[(rental_data["Year"] == 2019)]
rent_2020 = rental_data[(rental_data["Year"] == 2020)]
rent_2021 = rental_data[(rental_data["Year"] == 2021)]
rent_2022 = rental_data[(rental_data["Year"] == 2022)]
```

```python
rent_2018.head()
```

```python
rent_2018.columns
```

```python
rent_rate_2018 = rent_2018.groupby("Province")["Total "].mean()
rent_rate_2019 = rent_2019.groupby("Province")["Total "].mean()
rent_rate_2020 = rent_2020.groupby("Province")["Total "].mean()
rent_rate_2021 = rent_2021.groupby("Province")["Total "].mean()
rent_rate_2022 = rent_2022.groupby("Province")["Total "].mean()
```

```python
rent_rate_2018
```

```python
#Average rent by Province in 2022
rent_rate_2022 = rent_rate_2022.sort_values(ascending=False)
plt.bar(rent_rate_2022.index, rent_rate_2022.values)
plt.xlabel("Province")
plt.ylabel("Average Rent $")
plt.title("Average Rent Charged by Province in 2022")
plt.show()
```

```python
#get average rent/year for each region in Ontario 
ONrent_2022 = rent_2022[(rent_2022["Province"] == "Ont.")]
ONrent_2021 = rent_2021[(rent_2021["Province"] == "Ont.")]
ONrent_2020 = rent_2020[(rent_2020["Province"] == "Ont.")]
ONrent_2019 = rent_2019[(rent_2019["Province"] == "Ont.")]
ONrent_2018 = rent_2018[(rent_2018["Province"] == "Ont.")]
#rent_2022["Province"].unique()
```

```python
#Determine values for the Centres in Ontario
ONrent_2022["Centre"].unique()
```

```python
ONrent_2022
```

```python
#Get average rent by region in Ontario in 2022
ONCentre_rates2022 = ONrent_2022.groupby("Centre")["Total "].mean()
ONCentre_rates2022 = ONCentre_rates2022.sort_values(ascending=False)
ONCentre_rates2022
```

```python
#Plot combined average rent charged by municipality (Centre)
plt.barh(ONCentre_rates2022.index,ONCentre_rates2022.values)
plt.ylabel("Region")
plt.xticks(rotation=0)
plt.xlabel("Average Rent $")
plt.title("Average Rent Charged by Region In Ontario - 2022")
fig1 = plt.gcf()
plt.show()
fig1.savefig('ON_Region_rent.png',bbox_inches='tight', pad_inches=0.5)
```

```python
#Get average rent by zone in Toronto
Toronto_rates2022 = ONrent_2022[ONrent_2022["Centre"] == "Toronto"]
Toronto_rates2022
```

```python
Tor_zone_rates2022 = Toronto_rates2022.groupby("Zone")["Total "].mean()
Tor_zone_rates2022 = Tor_zone_rates2022.sort_values(ascending=False)
Tor_zone_rates2022
```

```python
#Plot bar graph of average rent chanrged within the GTA
plt.barh(Tor_zone_rates2022.index,Tor_zone_rates2022.values)
plt.ylabel("City")
plt.yticks(rotation=0)
plt.xlabel("Average Rent $")
plt.title("Average Rent Charged within the GTA - 2022")
fig1 = plt.gcf()
plt.show()
fig1.savefig('GTA_rent.png',bbox_inches='tight', pad_inches=0.5)
```

# 5 Year Trend 

```python
#Trend in rental rate for top 5 zONES (cities/town) in the GTA
zones_to_plot = ["Brampton (West)","Oakville","Mississauga (Northeast)","North York (Northeast)","North York (Northwest)"]

zone_rent_trend_df = rental_data[['Year','Zone','Total ']]

zone_rent_trend_df = zone_rent_trend_df[(zone_rent_trend_df['Zone'].isin(zones_to_plot))]

zone_rent_trend_df = pd.pivot_table(zone_rent_trend_df, index='Year', columns='Zone', values='Total ', aggfunc='mean', fill_value=0).reset_index()

zone_rent_trend_df.columns = ["Year", "Brampton (West)","Oakville","Mississauga (Northeast)","North York (Northeast)","North York (Northwest)"]
zone_rent_trend_df.head()
```

```python
#plot 5 year rent Trend for the 5 most expenseive regions (2022) in the GTA:
plt.plot( 'Year', 'Brampton (West)', data=zone_rent_trend_df, marker='o', markerfacecolor='skyblue', markersize=5, color='skyblue', linewidth=2)
plt.plot( 'Year', 'Oakville', data=zone_rent_trend_df, marker='o', markerfacecolor='red', markersize=5, color='red', linewidth=2)

plt.plot( 'Year', 'Mississauga (Northeast)', data=zone_rent_trend_df, marker='o', markerfacecolor='orange', markersize=5, color='orange', linewidth=2)
plt.plot( 'Year', 'North York (Northeast)', data=zone_rent_trend_df, marker='o', markerfacecolor='black', markersize=5, color='black', linewidth=2)
plt.plot( 'Year', 'North York (Northwest)', data=zone_rent_trend_df, marker='o', markerfacecolor='green', markersize=5, color='green', linewidth=2)

plt.xticks([2018,2019,2020,2021,2022])
plt.title("Trend - Average  Rent for Top 5 Zones in GTA (2018 - 2022)")
plt.ylabel("Average Rent $")
plt.xlabel("Year")
plt.grid()
plt.legend()

fig1 = plt.gcf()
plt.show()
fig1.savefig('Top 5.png', bbox_inches='tight', pad_inches=0.5)
```
